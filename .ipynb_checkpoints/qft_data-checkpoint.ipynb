```python
import numpy as np
from numpy.linalg import multi_dot
from scipy.linalg import norm
import matplotlib.pyplot as plt
%run -i qft.py
%run -i rn_cu_gadget.py
%run -i helper.py
```

```python
nflags=100
flag = 0
q1 = 1
q2 = 2
q3 = 3
q4 = 4
probs = np.linspace(0,0.015,150)
```

```python
qubits = [q1, q2]
state_dict = {(0,0,0):0, (0,1,0):1}
qft_circ = qft2_circ
res2, rej2, rejprob2 = qft_plot_data(state_dict, nflags, flag, qubits, qft_circ, probs, randuflag=True)
```

```python
plt.scatter(res2[0],res2[1], alpha=0.2, color='blue')
plt.scatter(rej2[0],rej2[1], alpha=0.2, color='red')
plt.xticks([0, 0.005, 0.010, 0.015],fontsize=16)
plt.xlabel(r'$p$', fontsize=20)
plt.yticks(fontsize=16)
plt.ylabel('Fidelity', fontsize=20)
plt.tight_layout()
```
